```python
import pandas as pd
import numpy as np

# Sayısal sütunlar
numeric_columns = red_wine.select_dtypes(include=[np.number])

# Korelasyon matrisi
correlation_matrix = numeric_columns.corr()

# Kendini tekrar eden verileri çıkarmak için
upper_triangle = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool))

# Büyükten küçüğe sıralama
high_corr = upper_triangle.stack().sort_values(ascending=False)

# En yüksek 10 korelasyon ikilisi
top_10_corr_pairs = high_corr.head(10)

print("Kırmızı Şarap En yüksek 10 korelasyon ikilisi: ")
for pair, corr_value in top_10_corr_pairs.items():
    print(f"{pair[0]} ve {pair[1]}: {corr_value:.2f}")
```

```python
from sklearn.preprocessing import StandardScaler

# Kırmızı şarap özelliklerini seçme
red_features = ['fixed_acidity_red', 'volatile_acidity_red', 'citric_acid_red', 'residual_sugar_red',
                'chlorides_red', 'free_sulfur_dioxide_red', 'total_sulfur_dioxide_red', 'density_red',
                'pH_red', 'sulphates_red', 'alcohol_red']

# Beyaz şarap özelliklerini seçme
white_features = ['fixed_acidity_white', 'volatile_acidity_white', 'citric_acid_white', 'residual_sugar_white',
                  'chlorides_white', 'free_sulfur_dioxide_white', 'total_sulfur_dioxide_white', 'density_white',
                  'pH_white', 'sulphates_white', 'alcohol_white']

# Özellikleri standartlaştırma
scaler_red = StandardScaler()
scaler_white = StandardScaler()

# Kırmızı şarap verilerini standartlaştırma
red_wine_scaled = scaler_red.fit_transform(red_wine[red_features])

# Beyaz şarap verilerini standartlaştırma
white_wine_scaled = scaler_white.fit_transform(white_wine[white_features])
```

```python
from sklearn.ensemble import RandomForestClassifier

# Kırmızı şarap için model oluşturma ve en önemli değişkeni bulma
model_red = RandomForestClassifier()
model_red.fit(red_wine_scaled, red_wine['quality_red'])

feature_importances_red = model_red.feature_importances_

plt.figure(figsize=(12, 6))
sns.barplot(x=red_features, y=feature_importances_red)
plt.title('Kırmızı şarap kalitesini en çok etkileyen özellikler')
plt.xticks(rotation=40)
plt.show()

# Beyaz şarap için model oluşturma ve en önemli değişkeni bulma
model_white = RandomForestClassifier()
model_white.fit(white_wine_scaled, white_wine['quality_white'])

feature_importances_white = model_white.feature_importances_

plt.figure(figsize=(12, 6))
sns.barplot(x=white_features, y=feature_importances_white)
plt.title('Beyaz şarap kalitesini en çok etkileyen özellikler')
plt.xticks(rotation=40)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

# Modeli oluşturma ve fit etme
model_red = LogisticRegression(max_iter=1000)
model_red.fit(red_wine_scaled, red_wine['quality_red'])

# Tahmin yapma
y_pred_red = model_red.predict(red_wine_scaled)

# Değerlendirme
print("Red Winex Classification Report:")
print(classification_report(red_wine['quality_red'], y_pred_red))
print('Accuracy:', accuracy_score(red_wine['quality_red'], y_pred_red))
```

```python
# Modeli oluşturma ve fit etme
model_white = LogisticRegression(max_iter=1000)
model_white.fit(white_wine_scaled, white_wine['quality_white'])

# Tahmin yapma
y_pred_white = model_white.predict(white_wine_scaled)

# Değerlendirme
print("White Wine Classification Report:")
print(classification_report(white_wine['quality_white'], y_pred_white))
print('Accuracy:', accuracy_score(white_wine['quality_white'], y_pred_white))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score


selected_columns = [
    'fixed_acidity_red', 'volatile_acidity_red', 'citric_acid_red',
    'residual_sugar_red', 'chlorides_red', 'free_sulfur_dioxide_red',
    'total_sulfur_dioxide_red', 'density_red', 'pH_red', 'sulphates_red',
    'alcohol_red', 'quality_red'
]

red_wine = red_wine[selected_columns]

red_wine = red_wine.dropna()

X = red_wine.drop('quality_red', axis=1)  # Özellikler
y = red_wine['quality_red']  # Hedef değişken

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Karar ağacı modelini oluşturma ve eğitme
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
